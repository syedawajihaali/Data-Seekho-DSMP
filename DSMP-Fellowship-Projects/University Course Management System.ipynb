##**Project Description:**
This project implements a Course Management System using object-oriented programming (OOP) principles in Python. It involves designing a system where various types of courses can be created, with specific attributes and behaviors, while leveraging key OOP concepts such as inheritance, encapsulation, polymorphism, and abstraction.

##**Project Objectives:**
* **Centralized Course Management:** Provide a platform for managing all course-related data.
* **Automated Enrollment:** Streamline student enrollment and course registration.
* **Syllabus Management:** Simplify the creation and distribution of course syllabi.
* **Scheduling:** Allow for structured management of lectures and lab sessions.
* **Custom Course Types:** Support multiple course types with specific attributes and schedules (e.g., Programming Fundamentals, DBMS).


##**CODE:**

```python
# Base class: Course
class Course:
    def __init__(self, course_name, course_code, credits, instructor, semester, department):
        self.__course_name = course_name  # Encapsulation (Private Attribute)
        self.__course_code = course_code
        self.__credits = credits
        self.__instructor = instructor
        self.__semester = semester
        self.__department = department

    # Public method to get the course syllabus
    def get_syllabus(self):
        return f"Syllabus for {self.__course_name}."

    # Public method to enroll a student in the course
    def enroll_student(self, student_name):
        return f"{student_name} has been enrolled in {self.__course_name}."

    # Getters for encapsulated attributes
    def get_course_name(self):
        return self.__course_name

    def get_course_code(self):
        return self.__course_code

    def get_credits(self):
        return self.__credits

    def get_instructor(self):
        return self.__instructor

    def get_semester(self):
        return self.__semester

    def get_department(self):
        return self.__department

    # Polymorphic method: This will be overridden in derived classes
    def get_schedule(self):
        return f"{self.__course_name} is scheduled for 3 lectures per week."

# Derived class: ProgrammingCourse
class ProgrammingFundamentals(Course):
    def __init__(self, course_name, course_code, credits, instructor, semester, department, programming_language, tools):
        super().__init__(course_name, course_code, credits, instructor, semester, department)
        self.programming_language = programming_language  # Additional attribute
        self.tools = tools  # Tools used in the course (e.g., IDEs, frameworks)

    # Method to get practical exercises in the programming language
    def get_practical_exercises(self):
        return f"Practical exercises in {self.programming_language}, using {self.tools}."

    # Polymorphism: Overriding the get_schedule method
    def get_schedule(self):
        return f"{self.get_course_name()} includes coding assignments, 2 lab sessions, and 1 theory lecture per week."

# Derived class: MathematicsCourse
class Calculus(Course):
    def __init__(self, course_name, course_code, credits, instructor, semester, department, math_field, software_tools):
        super().__init__(course_name, course_code, credits, instructor, semester, department)
        self.math_field = math_field  # Additional attribute
        self.software_tools = software_tools  # Additional attribute for software used (e.g., MATLAB)

    # Method to get a formula sheet for the math field
    def get_formula_sheet(self):
        return f"Formula sheet for {self.math_field} using {self.software_tools}."

    # Polymorphism: Overriding the get_schedule method
    def get_schedule(self):
        return f"{self.get_course_name()} includes 4 theory lectures and problem-solving sessions per week."

# Derived class: ScienceCourse
class Electronics(Course):
    def __init__(self, course_name, course_code, credits, instructor, semester, department, lab_hours, experiment_count):
        super().__init__(course_name, course_code, credits, instructor, semester, department)
        self.lab_hours = lab_hours  # Additional attribute
        self.experiment_count = experiment_count  # Number of experiments done in the course

    # Method to get lab experiment details
    def get_lab_details(self):
        return f"{self.get_course_name()} includes {self.experiment_count} experiments with {self.lab_hours} lab hours."

    # Polymorphism: Overriding the get_schedule method
    def get_schedule(self):
        return f"{self.get_course_name()} includes 3 lectures and 2 lab sessions per week."

# Derived class: DBMSCourse
class DBMSCourse(Course):
    def __init__(self, course_name, course_code, credits, instructor, semester, department, database_systems, lab_sessions):
        super().__init__(course_name, course_code, credits, instructor, semester, department)
        self.database_systems = database_systems  # Databases covered (e.g., MySQL, Oracle)
        self.lab_sessions = lab_sessions  # Number of lab sessions

    # Method to get the database systems covered
    def get_database_systems(self):
        return f"{self.get_course_name()} covers the following database systems: {', '.join(self.database_systems)}."

    # Method to get lab session details
    def get_lab_sessions(self):
        return f"{self.get_course_name()} includes {self.lab_sessions} lab sessions for hands-on database design and management."

    # Polymorphism: Overriding the get_schedule method
    def get_schedule(self):
        return f"{self.get_course_name()} includes 2 theory lectures and {self.lab_sessions} lab sessions per week."


# Displaying course information excluding enrollment
def print_course_info(course):
    print(f"{'=' * 40}")
    print(f"Course Name: {course.get_course_name()}")
    print(f"Course Code: {course.get_course_code()}")
    print(f"Credits: {course.get_credits()}")
    print(f"Instructor: {course.get_instructor()}")
    print(f"Department: {course.get_department()}")
    print(f"Semester: {course.get_semester()}")
    print(f"Syllabus: {course.get_syllabus()}")
    print(f"Schedule: {course.get_schedule()}")

    if isinstance(course, ProgrammingFundamentals):
        print(f"Practical Exercises: {course.get_practical_exercises()}")
    elif isinstance(course, Calculus):
        print(f"Formula Sheet: {course.get_formula_sheet()}")
    elif isinstance(course, Electronics):
        print(f"Lab Details: {course.get_lab_details()}")
    elif isinstance(course, DBMSCourse):
        print(f"Database Systems: {course.get_database_systems()}")
        print(f"Lab Sessions: {course.get_lab_sessions()}")
    print(f"{'=' * 40}\n")

# Separate method for enrolling students with different names
def enroll_in_course(course, student_name):
    print(course.enroll_student(student_name))


# Example usage
course1 = ProgrammingFundamentals("Programming Fundamentals", "CS101", 3, "Dr. Badar Sami", "Fall 2024", "CS Department", "Python", "IDLE")
course2 = Calculus("Calculus", "MATH201", 4, "Dr. Noor Fatima", "Fall 2024", "Mathematics Department", "Calculus", "MATLAB")
course3 = Electronics("Physics", "PHY101", 3, "Dr. Ali", "Fall 2024", "Physics Department", 4, 6)
course4 = DBMSCourse("Database Management Systems", "CS202", 4, "Dr. Nadeem Mehmood", "Spring 2024", "CS Department", ["MySQL", "Oracle"], 2)

# Printing course info
print_course_info(course1)
enroll_in_course(course1, "Wajiha")
print_course_info(course2)
enroll_in_course(course2, "Ali")
print_course_info(course3)
print_course_info(course4)
```

##**Code Description:**
* The system consists of a base class called **Course**, which has private attributes such as course_name, course_code, and students_enrolled.
* It includes methods like get_syllabus(), enroll_student(), and get_enrolled_students(), as well as getter methods for its private attributes.
* Several derived classes extend Course, including ProgrammingFundamentals, Calculus, Electronics, and DBMSCourse.
* Each derived class adds unique attributes and overrides the get_schedule() method to provide specific course schedules.

##**OOP Concepts Used:**
* **Encapsulation** is implemented through private attributes in the Course class, ensuring that critical course details are not directly modified, promoting data security.
* **Inheritance** allows derived classes to inherit common attributes and methods while adding specific features tailored to each course type, such as programming languages for ProgrammingFundamentals.
* **Polymorphism** is demonstrated as the get_schedule() method is overridden in each derived class, providing specialized behaviors for different courses.
* **Abstraction** allows the Course class to handle general course details while derived classes focus on specific implementations, enabling higher-level interactions with the course system.
